```python
# import libraries
import pandas as pd
```

```python
df = pd.read_csv("dataset/katalog_gempa.csv")
df
```

```python
# drop columns not needed
df = df.drop(columns=["strike1", "dip1", "rake1", "strike2", "dip2", "rake2"])
df
```

```python
# change column names to english
df = df.rename(columns={"tgl" : "date", "ot" : "timestamp", "lat" : "latitude", "lon" : "longitude", "mag" : "magnitude", "remark" : "location"})
df
```

```python
# convert date to datetime format
df["date"] = pd.to_datetime(df.date)
df["timestamp"] = pd.to_datetime(df.timestamp).dt.time
df
```

```python
# select column
location = df["location"]

# remove "Indonesia" part in column
def removeSuffix(loc):
  new_loc = ""
  if " - Indonesia" in loc:
    new_loc = loc.replace(" - Indonesia", "")
    return new_loc
  else:
    return loc

new_location = location.apply(removeSuffix)

df["new location"] = new_location
df
```

```python
df = df.drop(columns=["location"])
df = df.rename(columns={"new location" : "location"})
df
```

```python
df.to_csv("dataset/earthquake_clean.csv", encoding='utf-8', index=False)
```
